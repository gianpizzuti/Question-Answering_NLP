```python
import pyterrier as pt

if not pt.started():
  pt.init()
```

```python
import pandas as pd

# Load the JSON file into a pandas DataFrame
df = pd.read_json('medical_meadow_wikidoc_medical_flashcards.json')
df
```

```python
#deleting teh last 4 rows since they are empty
df = df[:-4]
df
```

```python
#insert the docno column for further pyterrier motivations
df.insert(0, 'docno', 'd' + (df.index + 1).astype(str))
df
```

```python
#indexing
indexer = pt.DFIndexer("./index_docs", overwrite=True)
index_ref = indexer.index(df["instruction"], df["input"], df["output"],df["docno"])
index_ref.toString()
```

```python
index = pt.IndexFactory.of(index_ref)
```

```python
#some statics on the documents
print(index.getCollectionStatistics().toString())
```

### Searching an Index

Now that we have indexed our documents, we can run a search over the document collection:

```python
#indexing documents by TF
query = "Mg2+"

tf = pt.BatchRetrieve(index, wmodel="Tf")
tf.search(query)
```

```python
#indexing documents by TF-IDF
query = "Mg2+"

tf_idf = pt.BatchRetrieve(index, wmodel="TF_IDF")
tf_idf.search(query)
```

```python
#indexing documents by BM25
query = "Mg2+"

bm25 = pt.BatchRetrieve(index, wmodel="BM25")
bm25.search(query)
```

COMPOSITION OF RETRIEVERS

```python
#Tf and TF-IDF
query = "Mg2+"

composition = tf >> tf_idf

composition.search(query)
```

```python
#Tf and BM25
query = "Mg2+"

composition = tf >> bm25

composition.search(query)
```

```python
#TF-IDF and BM25
query = "Mg2+"

composition = tf >> bm25

composition.search(query)
```

UNION OF RETRIEVIERS

```python
#combine the top 10 results from TF with the top 10 results from TF-IDF
union = (tf % 10) | (tf_idf % 10)

union.search(query)
```

```python
#add som re-ranking to our pipeline using BM25
pipeline = ((tf % 10) | (tf_idf % 10)) >> bm25

pipeline.search(query)
```
